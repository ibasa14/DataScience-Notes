Una vez vistas las bases de la librería, sus funcionalidades más sencillas y el tipo de gráfico más simple, es momento de entrar más en profundidad y hablar sobre los otros 3 tipos más utilizados:
- scatter()
- contour() / contourf()
- hist()

#### Gráfico de dispersión
Es un tipo de gráfico similar al visto durante el comienzo de la guía. A diferencia de éste, los puntos que se dibujan son al menos por defecto, independientes. Es decir, no vienen unidos por segmentos. A continuación vamos a pintar el mismo conjunto de datos con ambas funciones para ver claramente la diferencia:

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x = np.linspace(1, 10, 30)

figure, ax = plt.subplots(1,2, figsize = (10, 4))
ax[0].plot(x, np.sin(x))
ax[1].scatter(x, np.sin(x))

ax[0].set(title = 'plot')
ax[1].set(title = 'scatter');
```

Esta función presenta gran capacidad de personalización, podemos cambiar parámetros como el color, el tipo de marcador, el tamaño del marcador, la transparencia (**alpha**), etc. A continuación las propiedades con el nombre del parámetro:

- tamaño: **s**
- color: **color**
- tipo de marcador: **marker**. Algunos ejemplos: '.', ',', '*', 'x', 'v', 's', 'o', etc.
    
Vamos a ver un ejemplo con diferentes combinaciones:

```python
x = np.linspace(1, 10, 30)

figure = plt.figure(figsize = (8, 4))
plt.scatter(x, np.sin(x), marker = 's', color = 'red', s = 5, label = 'Cuadrados')
plt.scatter(x, np.sin(x-2), marker = 'o', color = 'green', s = 15, label = 'Discos')
plt.scatter(x, np.sin(x-4), marker = 'v', color = 'blue', s = 35, label = 'Triángulos')

plt.legend();
```

#### Gráfico de contorno

Son muy útiles para mostrar **datos tridimensionales** en dos dimensiones. Tenemos fundamentalmente dos opciones:

- contour()
- contourf() : con relleno de color

Veamos algunos ejemplos:

```python
# definimos los puntos sobre el eje X e Y (vectores de 1 dimensión)
x = np.linspace(0, 5, 5)   
y = np.linspace(0, 5, 5)

# Creamos las matrices de coordenadas con meshgrid  (matrices de 2 dimensiones)
X, Y  = np.meshgrid(x, y)

# Generamos los valores Z (los que vamos a mostrar con colores)
Z = np.sin(X) +  np.cos(Y/2)    # función inventada para el ejemplo

figure, ax = plt.subplots(1,2, figsize = (10, 4))
ax[0].contour(X, Y, Z)
ax[1].contourf(X, Y, Z)

ax[0].set(title = 'contour')
ax[1].set(title = 'contourf');
```

Al igual que en los gráficos anteriores, en contour podemos especificar los colores que queremos visualizar (en este caso gama de colores). Para ello usamos el parámetro **cmap**. Algunas familias de cmap ejemplo:

- 'spring', 'summer', 'autumn', 'winter'
- 'Greys', 'Purples', 'Blues', 'Greens', 'Oranges'
- 'GnBu', 'PuBu', 'YlGnBu'



```python
x = np.linspace(0, 5, 5)   
y = np.linspace(0, 5, 5)

X, Y  = np.meshgrid(x, y)

Z = np.sin(X) +  np.cos(Y/2)    

figure, ax = plt.subplots(1,2, figsize = (10, 4))
ax[0].contourf(X, Y, Z, cmap = 'spring')    # especificamos los valores de cmap
ax[1].contourf(X, Y, Z, cmap = 'PuBu')

ax[0].set(title = 'contourf spring')
ax[1].set(title = 'contourf PuBu');

```

Además podemos elegir si queremos ver la barra de colores para que nos sea más sencillo seguir los valores que esta tomando el gráfico. Basta con especificar una vez creado el gráfico: **plt.colorbar()**.

```python
plt.contourf(X, Y, Z, cmap = 'Greens')
plt.colorbar();
```

#### Histogramas

Un histograma se presenta como la mejor opcion para comprender rápidamente la distribución de un conjunto de datos. En este apartado vamos a ver como mostrarlos en esta librería y algunos de los parámetros que pueden ajustarse:

```python
# creamos la variables que queremos estudiar

# distribución normal, media = 0, desviacion estandar = 10, n = 1000
normal = np.random.normal(0, 10, 1000)

# bins = numero de agrupaciones/divisiones. Por defecto = 10
plt.hist(normal, bins = 20);
```

Si queremos mostrar dos distribuciones en el mismo grafico, podemos definir un color para cada uno, y un cierto valor de transparencia (**alpha**) para que se vean todos en caso de superponerse:

```python
normal = np.random.normal(0, 10, 1000)

# distribucion uniforme, valor minimo = -30, maximo = 30, n = 1000
uniforme = np.random.uniform(-30, 30, 1000)

# bins: número de divisiones/agrupaciones del histograma
plt.hist(normal, bins = 20, color = 'green', alpha = 0.2, label = 'normal');

#alpha para definir cierta transparencia
plt.hist(uniforme, bins = 20, color = 'blue', alpha = 0.2, label = 'uniforme');

plt.legend();
```

La función hist() nos da la psibilidad de **normalizar el histograma**, algo que es bastante común utilizar. El parámetro es **normed**.

```python
normal = np.random.normal(0, 10, 1000)

plt.hist(normal, bins = 20, color = 'blue', alpha = 0.2,\
         density = True, label = 'normalizada');

plt.legend();
```
