## Graficamos

A continuación, en este caítulo vamos a ver las funcionalidades más básicas que ofrece la librería


```python
import matplotlib.pyplot as plt
import numpy as np

#durante esta guía usaremos este por comodidad
%matplotlib inline  

x = np.linspace(0, 10, 200)  # array de 200 valores equiespaciados en rango 0-10
plt.plot(x, np.sin(x))
```

Si quisieramos gráficar una función más en el mismo plt:

```python
x = np.linspace(0, 10, 200)

plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
```

En este caso, si mostramos dos conjuntos de datos en el mismo plt, es muy recomendable incorporar una leyenda:

Para ellos primero debemos especificar una etiqueta (**label**) para cada función y después llamar a **plt.legend()**

```python
x = np.linspace(0, 10, 200)  

plt.plot(x, np.sin(x), label = 'seno')
plt.plot(x, np.cos(x), label = 'coseno')
plt.legend()
```

En este caso, la leyenda esta adoptando el formato por defecto y se esta posicionando donde el propio matplotlib considera más oportuno. 

Podemos dar diferentes formatos a las curvas, por ejemplo **tipo de línea** y **color**. Ejemplo a continuación:


```python
x = np.linspace(0, 10, 200)  

plt.plot(x, np.sin(x), label = 'curva_1', color = 'blue', linestyle = '-')
plt.plot(x, np.cos(x), label = 'curva_2',  color = 'red', linestyle = '-.')
plt.plot(x, np.cos(x+2), label = 'curva_3',  color = '#FFDD44', linestyle = '--')

plt.legend();
```

A lo mejor nos interesa crear una malla y dentro de ella dibujar diferentes gráficos. 

En este caso tenemos que hacer uso de la función **subplots()**. Esta función toma como dos primeros parámetros el número de filas de la malla y el número de columnas.

Devuelve una tupla formada por los objetos **Figure** y una lista de **Axes**, tantos como se haya indicado en la malla.

Figure representa la figura completa (la malla matplotlib básicamente). Axes representa el contenedor de cada uno de los gráficos (lo que vemos).

Un ejemplo:

```python
x = np.linspace(0, 10, 200)

# creo una malla con 1 fila y dos columnas
figure, ax = plt.subplots(1, 2)

ax_0 = ax[0].plot(x, np.sin(x)) #primer axes (primera fila primera columna)
ax_1 = ax[1].plot(x, np.cos(x)) #segundo axes  (primera fila segunda columna)
```

Podemos también dibujar varias curvas en cada gráfico, poner leyenda en cada uno, etc.

Por otro lado, podemos establecer el tamaño de la figura cuando llamamos a la función subplots():

Ahora podríamos cargar esta imagen haciendo uso del objeto Image:

```python
x = np.linspace(0, 10, 200)

# modificamos el tamaño del objeto Figure con figsize
figure, ax = plt.subplots(1, 2, figsize = (10, 4))

ax_0_a = ax[0].plot(x, np.sin(x), label = 'curva_1', color='red') 
ax_0_b = ax[0].plot(x, np.cos(x), label = 'curva_2', color='blue') 
ax[0].legend() # incluimos leyenda al primer axes

ax_1 = ax[1].plot(x, np.sin(x)) 
```

### Título y nombre de ejes

Resulta interesante ponerle a cada gráfico un título diferente para poder diferenciarlos y también, como no, especificar el nombre de cada una de las variables del gráfico (ejes). Para esto hacemos uso de la función **set()**.

Vamos a ver un ejemplo en el que le ponemo al gráfico derecho de la malla ests nombres. Además aprovechamos para introducir un cierto espaciado entre cada uno de los gráficos de la malla. A través de la función **subplots_adjust** sobre el objeto Figure creado podemos especificar **márgenes** a cada lado e **interespaciado** entre gráficos:

```python
x = np.linspace(0, 10, 200)

figure, ax = plt.subplots(1, 2, figsize = (10, 4))

ax_0_a = ax[0].plot(x, np.sin(x), label = 'curva_1', color='red') 
ax_0_b = ax[0].plot(x, np.cos(x), label = 'curva_2', color='blue') 
ax[0].legend() 

ax_1 = ax[1].plot(x, np.sin(x))

# Especificamos estos 3 parámetros en el gráfico del lado derecho
ax[1].set(title = 'Gráfico lado derecho', xlabel = 'Nombre eje X', ylabel = 'Nombre eje Y')

figure.subplots_adjust(wspace=0.6) # wspace: interespaciado horizontal;  hspace: interespaciado vertical
```
