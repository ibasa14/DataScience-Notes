## ¿Qué es?

Matplotlib es una biblioteca de visualización de datos multiplataforma construida sobre NumPy.

Una de las características más importantes de Matplotlib es que funciona funcionará independientemente del sistema operativo que se esté utilizando o del formato de salida que desee.



## Contexto

Antes de ponernos a ver los diferentes tipos de gráficos, personalización y configuracion de los mismos, vamos a aclarar primero algo que muchas veces da lugar a confusiones...

La utilización de Matplotlib **depende del contexto**. ¿De qué contextos hablámos? 

- script (.py)
- terminal IPython 
- cuaderno IPython (.ipynb)

### Script

Si está utilizando Matplotlib desde dentro de un script, necesitarás hacer uso de la función **show ()** : inicia un ciclo de eventos, busca todos los objetos de figura actualmente activos y abre una o más ventanas interactivas que muestran las figuras.

Es importante tener en cuenta que esta función solo debe usarse 1 vez por script. Lo más normal es hacerlo al final una vez esten configurados todos los graficos (**plt** por convenio al hablar de matplotlib)

Entonces, por ejemplo, si queremos mostrar un gráfico a partir de un archivo .py podriamos hacer lo siguiente:

```python
# ------- archivo: dibuja.py ------

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)

plt.plot(x, np.sin(x))

plt.show()

# --------------------------------

# ------- terminal IPython  ------
#
#       $ python dibuja.py
#
#---------------------------------
```

### Terminal IPython

IPython necesita especifica el modo Matplotlib para ejecutarse correctamente en diferentes sistemas operativos. Para habilitar este modo, escribimos: **%matplotlib**.

Cualquier plt que se escriba se mostrará automáticamente. Si dibujamos un plt y queremos después cambiar algun parámetro y actualizarlo, usaremos la función: **draw()**.

### Cuaderno IPython

Tenemos varias opciones. La primera es hacer uso del comando %matplotlib y funciona de manera similar a la terminal IPython. 

La segunda es enbeber los plt dentro del cuaderno:
- %matplotlib notebook: dara lugar a plts **interactivos**
- %matplotlib inline: datá lugar a plts **estáticos**

```python
#estático
%matplotlib inline

x = np.linspace(0, 10, 100)

fig = plt.figure()
plt.plot(x, np.cos(x), '--');
```

```python
#dinámico
%matplotlib notebook

x = np.linspace(0, 10, 100)

fig = plt.figure()
plt.plot(x, np.cos(x), '--');
```
